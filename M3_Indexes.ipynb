# Smoothing of the Spectra

```python
#Trnasposing data to ingest in Spec1d2
tM3_cube=np.transpose(M3_cube.data,(2,1,0))

#Filtering with a 1d gaussian fucntion
spec1d2=Spectrum1D(spectral_axis=wavelengths[0::]*u.AA, flux=tM3_cube*u.dimensionless_unscaled)
spec1d2_gsmooth=gaussian_smooth(spec1d2, stddev=2)

#Slicing the cube to avoid the affected data at the longer wavelength 
tM3_cube_smooth=spec1d2_gsmooth.flux.value
slice1=tM3_cube_smooth[:,:,0:79]
slice2=tM3_cube[:,:,79:83]
spec1d2_final=np.dstack((slice1,slice2))

#Savind the data in a cube
M3_Gaussfilter=M3_cube.copy()
M3_Gaussfilter.data=np.transpose(spec1d2_final,(2,1,0))
x,y,z=M3_Gaussfilter.shape
plt.imshow(M3_Gaussfilter[5,:,:])
```

```python
#Fourier filtration cube
M3_Foufilter=M3_Gaussfilter.copy()
stack=[]

rowsf, colsf =M3_Gaussfilter[5,:,:].shape
maskf = np.ones((rowsf, colsf, 2), np.uint8)

cv2.rectangle(maskf, (0,((y//2)-10)), (((z//2)-30),((y//2)+10)), 0, -1)#Horizontals
cv2.rectangle(maskf, (((z//2)+30),((y//2)-10)), (z,((y//2)+10)), 0, -1)

for band in range(M3_Gaussfilter.data.shape[0]):
    
    imput_imgf=M3_Gaussfilter.data[band,:,:]
    fouxrasterf=cv2.dft(imput_imgf, flags=cv2.DFT_COMPLEX_OUTPUT)# Fourier transform 
    fouxraster_shiftf=np.fft.fftshift(fouxrasterf)

    mfouxrasterf=fouxraster_shiftf*maskf #Mask
    m_ishiftf= np.fft.ifftshift(mfouxrasterf)
    return_rasterf=cv2.idft(m_ishiftf)/(y*z)
    return_rasterf= cv2.magnitude(return_rasterf[:, :, 0], return_rasterf[:, :, 1])
    stack.append(return_rasterf)

M3_Foufilter.data=np.array(stack)

#Plot to check
fig = plt.figure(figsize=(5,5))

ax1 = fig.add_subplot(1,2,1)
plt.imshow(M3_Gaussfilter[5,:,:], cmap="gray")

ax2 = fig.add_subplot(1,2,2)
plt.imshow(M3_Foufilter[5,:,:], cmap="gray")
```

```python
#Fourier filtration function2d
def fourier_filter(imput_data):
    rows,cols=imput_data.shape
    mask= np.ones((rows, cols, 2), np.uint8)
    y,z=imput_data.shape

    cv2.rectangle(mask, (0,((y//2)-10)), (((z//2)-30),((y//2)+10)), 0, -1)#Horizontals
    cv2.rectangle(mask, (((z//2)+30),((y//2)-10)), (z,((y//2)+10)), 0, -1)
 
    imput_img=imput_data.data
    fouraster=cv2.dft(imput_img, flags=cv2.DFT_COMPLEX_OUTPUT)# Fourier transform 
    fouraster_shift=np.fft.fftshift(fouraster)

    mfouraster=fouraster_shift*mask #Mask
    m_ishift= np.fft.ifftshift(mfouraster)
    return_raster=cv2.idft(m_ishift)/(y*z)
    return_raster= cv2.magnitude(return_raster[:, :, 0], return_raster[:, :, 1])
    
    imput_data.data=np.array(return_raster)
    return imput_data
```

```python
#Continumm functions
#Function for 1000
def continnum_1000(spectra_cube, spectra_hull):
            shoulder_0p=np.where(spectra_hull[0:20] == max(spectra_hull[0:20]))[0][-1]
            shoulder_0y=spectra_cube[shoulder_0p]
            shoulder_0x=wavelengths[shoulder_0p]
            shoulder_1p=np.where(spectra_hull[20:40] == max(spectra_hull[20:40]))[0][-1]+20
            shoulder_1y=spectra_cube[shoulder_1p]
            shoulder_1x=wavelengths[shoulder_1p]
            interpolation_x=np.array(shoulder_0x)
            interpolation_x1=np.append(interpolation_x,shoulder_1x)
            interpolation_y=np.array(shoulder_0y)
            interpolation_y1=np.append(interpolation_y,shoulder_1y)
            fit_1000=np.polyfit(interpolation_x1,interpolation_y1,1)
            return fit_1000

#Function for 2000
def continnum_2000(spectra_cube, spectra_hull):
            shoulder_2p=np.where(spectra_hull[40:66] == max(spectra_hull[40:66]))[0][-1]+40
            shoulder_2y=spectra_cube[shoulder_2p]
            shoulder_2x=wavelengths[shoulder_2p]
            shoulder_3p=np.where(spectra_hull[66:76] == max(spectra_hull[66:76]))[0][-1]+66
            shoulder_3y=spectra_cube[shoulder_3p]
            shoulder_3x=wavelengths[shoulder_3p]
            interpolation_x2=np.array(shoulder_2x)
            interpolation_x3=np.append(interpolation_x2,shoulder_3x)
            interpolation_y2=np.array(shoulder_2y)
            interpolation_y3=np.append(interpolation_y2,shoulder_3y)
            fit_2000=np.polyfit(interpolation_x3,interpolation_y3,1)
            return fit_2000
```
