```python
def OMAT (cube):
    
    mean_750 = np.mean(cube[5:8, :, :], axis=0)
    mean_950 = np.mean(cube[15:18, :, :], axis=0)

    OMAT= ((cube[6,:,:] - 0.08)**2 + ((mean_950/mean_750)-1.18)**2)**0.5

    return OMAT
M3_OMAT= OMAT(M3_cube)
M3_OMAT.rio.to_raster(fr'C:\Users\javie\Javier\Padova\Carpatus_Mons\Data\Carpatus\M3\MoonIndex\M3_OMAT.tif')
```

```python
def titanium (cube):
    '''Creates the TiO index. Which returns titanium oxide abundance in percentage.
    
    Input:
    gauss_cube = filtered cube.
    
    Output: 
    TiO index.'''

    mean_750 = np.mean(cube[5:8, :, :], axis=0)
    
    ti=cube[0,:,:].copy()
    tip=np.arctan2(((1.64*cube[0,:,:])/(mean_750*2.13)) - 0.42, (2.13*mean_750) - 0.00)
    ti.data=3.708 * (tip**5.979)
    return ti
M3_Ti= titanium(M3_cube)
M3_Ti.rio.to_raster(fr'C:\Users\javie\Javier\Padova\Carpatus_Mons\Data\Carpatus\M3\MoonIndex\M3_titanium.tif')
```

```python
def clementine (gauss_cube):
    '''Creates the clementine-like index. 
    
    Input:
    gauss_cube = filtered cube.
    
    Output:
    The clementine-like RGB composite.'''
    
    clem=gauss_cube[0:3,:,:].copy() 
    #Selecting the band to divide the filtered cube
    B1=gauss_cube[6,:,:]/gauss_cube[0,:,:]  
    B2=gauss_cube[6,:,:]/gauss_cube[19,:,:]
    B3=gauss_cube[0,:,:]/gauss_cube[6,:,:]
    clem.data=np.dstack((B1,B2,B3)).transpose(2,0,1)
    clem.data[clem.data > 3]=0
    return clem
```

```python
def iron (M3_cube):
    '''Creates the FeO index. Which return iron oxide abundance in percentage. 
    
    Input:
    gauss_cube = filtered cube.
    
    Output: 
    FeO index.'''
    
    mean_750 = np.mean(M3_cube[5:8, :, :], axis=0)
    mean_950 = np.mean(M3_cube[15:18, :, :], axis=0)
    
    fe=M3_cube[0,:,:].copy()
    fep = -np.arctan2(((1.94*mean_950) / (2.13*mean_750)) - 1.18, (2.13*mean_750) - 0.08)
    fe.data=(17.427*fep) -7.65
    return fe

M3_Fe= iron(M3_gauss)
M3_Fe.rio.to_raster(fr'C:\Users\javie\Javier\Padova\Carpatus_Mons\Data\Carpatus\M3\MoonIndex\M3_Fe.tif')
```

```python
def iron2(M3_cube):
    '''Creates the FeO index. Returns FeO abundance map (wt.%).'''

    mean_750 = np.mean(M3_cube[5:8, :, :], axis=0)
    mean_950 = np.mean(M3_cube[15:18, :, :], axis=0)
    
    ratio = mean_950 / mean_750
    fep = -np.arctan2(ratio - 1.18, mean_750 - 0.08)
    
    fe = M3_cube[0, :, :].copy()
    fe.data = 10.59 * (fep ** 1.209)

    # --- Imprimir valores para los primeros 5x5 pxeles ---
    print("x, y, mean_950, mean_750, ratio, fep")
    for i in range(5):
        for j in range(5):
            print(i, j,
                  float(mean_950[i, j]),
                  float(mean_750[i, j]),
                  float(ratio[i, j]),
                  float(fep[i, j]))
    
    return fe
M3_Fe2= iron2(M3_gauss)
M3_Fe2.rio.to_raster(fr'C:\Users\javie\Javier\Padova\Carpatus_Mons\Data\Carpatus\M3\MoonIndex\M3_Fe2.tif')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def find_origin(M3_cube, plot=True):
    """
    Encuentra el 'probable origin' (x0, y0) del cubo completo,
    basado en el scatter plot entre R750 y R950/R750.
    """
    
    # Bandas promediadas
    mean_750 = np.mean(M3_cube[5:8, :, :], axis=0)
    mean_950 = np.mean(M3_cube[15:18, :, :], axis=0)
    
    # Calcular ratio
    ratio = mean_950 / mean_750
    
    # Aplanar
    x = mean_750.ravel()
    y = ratio.ravel()
    
    # Quitar NaNs e infinitos
    mask = np.isfinite(x) & np.isfinite(y)
    x, y = x[mask], y[mask]
    
    # Distancia al origen (0,0)
    dist = np.sqrt(x**2 + y**2)
    
    # ndice del punto ms cercano
    idx = np.argmin(dist)
    
    # Extraer ese punto
    x0 = x[idx]
    y0 = y[idx]
    
    print(f"Punto ms cercano al origen (x0, y0): ({x0:.5f}, {y0:.5f})")
    
    # Graficar (opcional)
    if plot:
        plt.figure(figsize=(6,6))
        plt.scatter(x, y, s=1, alpha=0.3, label='All pixels')
        plt.scatter(x0, y0, color='yellow', s=80, marker='*', label='Closest to origin')
        plt.xlabel("R750")
        plt.ylabel("R950 / R750")
        plt.title("Closest Point to Origin from Full Cube")
        plt.legend()
        plt.show()
    
    return x0, y0

find_origin(M3_cube_initial.data)

find_origin (M3_cube_initial.data)
```
