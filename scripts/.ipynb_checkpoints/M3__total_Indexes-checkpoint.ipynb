# Obtaining all the indexes with one function

The next functions output the final cubes with all the indexes for the two continuum-removal methods. In other notebooks, the procces is described in detail and each product is created step by step.

```python
#Obtaining all the indexes in a single cube by the convex hull method. CAUTION. n_jobs paralelyze. Use -1 for all cores, 
#or a negative number to leave some cores free (e.g., -2 means all but 2 cores).If None, defaults to 1 (no parallelism).
M3_full_CH=MoonIndex.indexes.indexes_total_CH(M3_cube,wavelengths_full, n_jobs=None)
M3_full_CH.rio.to_raster('M3_full_CH.tif')
```

```python
#Obtaining all the indexes in a single cube by the convex second-and-first-order fit method. CAUTION. n_jobs paralelyze. Use -1 for all cores, 
#or a negative number to leave some cores free (e.g., -2 means all but 2 cores).If None, defaults to 1 (no parallelism).
M3_full_SAFO=MoonIndex.indexes.indexes_total_SAFO(M3_cube,wavelengths_full,2,1, n_jobs=None)
M3_full_SAFO.rio.to_raster('M3_full_SAFO.tif')
```

```python
#Plotting some random indexes to check, you can see the documentation to see possible combination
M3_full_CHa=M3_full_CH.to_array() #Tranforms the dataset to a dataarray to plot
MoonIndex.plotting.cube_plot(M3_full_CH[3:6,:,:],5,
                             'M3 indexes CH')
```
